```python
def add (a,b):
  #This is  a function of addition
  """This is  a function of addition a & b"""
  return a+b
def sub (a,b):
  """This is  a function of substruction a & b"""
  return a-b
def mul (a,b):
  """This is  a function of multiplication a & b"""
  return a*b
def div (a,b):
  """This is  a function of divition a & b"""
  return a/b
def floor_div (a,b):
  """This is  a function of floor_divition a & b"""
  return a//b
def power (a,b):
  """This is  a function of power a & b"""
  return a**b
def reminder (a,b):
  """This is  a function of reminder a & b"""
  return a%b
a=float (input('Enter The value 01 :'))
b=float(input('Enter the value 02 :'))
print(add.__doc__)
print(a,"+" ,b,"=" ,add(a,b))
print(f'{a} + {b} "=" ,{add(a,b)}')
print(sub.__doc__)
print(a,"-" ,b,"=" ,sub(a,b))
print(f'{a} - {b} "=" ,{sub(a,b)}')
print(mul.__doc__)
print(a,"*" ,b,"=" ,mul(a,b))
print(f'{a} * {b} "=" ,{mul(a,b)}')
print(div.__doc__)
print(a,"/" ,b,"=" ,div(a,b))
print(f'{a} / {b} "=" ,{div(a,b)}')
print(floor_div.__doc__)
print(a,"//" ,b,"=" ,floor_div(a,b))
print(f'{a} // {b} "=" ,{floor_div(a,b)}')
print(power.__doc__)
print(a,"^" ,b,"=" ,power(a,b))
print(f'{a} ^ {b} "=" ,{power(a,b)}')
print(reminder.__doc__)
print(a,"%" ,b,"=" ,reminder(a,b))
print(f'{a} % {b} "=" ,{reminder(a,b)}')
```

```python
def checkLeap(year):
  if ((year%4==0) and
    (year%100!=0) or
    (year%400==0)):
    print('leap year')
  else:
    print('not leap year')
year = int (input ("Enter your year"))
checkLeap(year)
```

```python
def is_leap_year(year):
    return year % 4 == 0 and (year % 100 != 0 or year % 400 == 0)
year = int(input("Enter a year: "))
if is_leap_year(year):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")
```
